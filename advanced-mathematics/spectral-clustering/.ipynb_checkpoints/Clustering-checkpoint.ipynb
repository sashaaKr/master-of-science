```python
def create_2d_graph(nof_nodes):
    G = np.zeros((nof_nodes,nof_nodes), dtype=int)
    size_grp1 = nof_nodes/2
    size_grp1 = size_grp2
    G[:size_grp1,:size_grp1]=1
    G[size_grp2:,size_grp2:]=1
    nof_common_nodes = size_grp1/3
    G[5,18]=1
    G[18,5]=1
    G[4,19]=1
    G[19,4]=1
    print ("G")
    print (G)
    edges = []
    for i in range(nof_nodes):
        for j in range (nof_nodes):
            if(G[i][j]==1):
                tmp = (i,j)
                edges.append(tmp)
                
    return G,edges
```

```python
np.set_printoptions(threshold=sys.maxsize)

nof_nodes = 20
radius = 0.5

# G = nx.random_geometric_graph(nof_nodes, radius)
A,edges = create_2d_graph(nof_nodes)
# print ("edges")
# print (edges)
G = nx.Graph()  # or DiGraph, MultiGraph, MultiDiGraph, etc
G.add_edges_from(edges)  # using a list of edge tuples
#G=nx.from_numpy_matrixG(edges)
# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(G, "pos")



# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(G, ncenter))

L = nx.laplacian_matrix(G)
#A = nx.convert_matrix.to_numpy_array(G, dtype=int)
DergeeList = list(G.degree)
D = np.zeros((nof_nodes, nof_nodes),dtype=int)

for node in range(len(DergeeList)):
    D[node][node] = int(DergeeList[node][1])
Lap = D - A

options = {"node_color": "C0", "node_size": 100}


labels = {}    
for node in G.nodes():
    labels[node] = node
```

```python
def calculate_egein(norm_Lapl):
    val, vec = np.linalg.eigh(norm_Lapl)
    idx = np.argsort(val)
    val = val[idx]
    vec = vec[:,idx]
#     print (val[0:1])
#     print (vec[0:1])
    return val,vec
```

```python
L_norm = nx.normalized_laplacian_matrix(G)
evals,evecs = calculate_egein(Lap)
second_vector = evecs[0]
print ("evals")
print (evals)
label_vector = np.zeros(len(second_vector))
median = np.median(second_vector)
print ("median: ")
print (median)
label_vector[second_vector>median]=1
label_vector[second_vector<=median]=0
color_map = []
for i in label_vector:
    if i == 0:
        color_map.append('red')
    else:
        color_map.append('green')
print ("label vector")
print (label_vector)

```
